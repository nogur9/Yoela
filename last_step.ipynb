```python
for key, items in trans.items():
    
    for item in items:
        if item not in df.columns:
            df[item] = False
            
        df.loc[df[key] == True, item] = True

    df = df.drop(key, axis=1)
```

```python
df.to_excel("survey_encoded.xlsx", index=0)
```
