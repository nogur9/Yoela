```python
import pandas as pd

df = pd.read_excel("survey_long_format.xlsx")
```

```python
cluster_order = {
    '7–10': 9,
    '1–3': 2,
    '4–6': 5
}

df["סוג הרשות"] = df["סוג הרשות"].map(authority_order)

authority_order = {
    "עיר גדולה (מעל 200 אלף תושבים)"
    :4,
    
    "עיר בינונית (50-100 אלף תושבים"
    :3,
    
    "עיר קטנה (20-50 אלף תושבים)"
    :2,

    "מועצה מקומית (עד 20 אלף תושבים)": 
    1, 

    "מועצה אזורית": None
}

df["אשכול חברתי-כלכלי"] = df["אשכול חברתי-כלכלי"].map(cluster_order)
```

```python
gap_order = {
    "במידה מועטה":1,
    "במידה בינונית": 2,
    "במידה גבוהה": 3,
    "במידה חריגה מאוד" :4
    
}

df["פער מול תקינה"] = df[
    "פער מול תקינה"
].map(gap_order)
```

```python
categorical_cols = [
    "מאפייני אוכלוסייה",
    "תחומי העומס"
]

df_encoded = pd.get_dummies(
    df,
    columns=categorical_cols,
    prefix=categorical_cols
)
```

```python
df_encoded.to_excel("survey_encoded.xlsx", index=False)
```
