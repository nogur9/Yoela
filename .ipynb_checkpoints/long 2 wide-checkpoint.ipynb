```python
import pandas as pd

# קריאת הדאטה
df = pd.read_excel("תשובות לשאלון.xlsx")

# משתנים שנשארים קבועים לכל משיב
id_vars = [
    "סוג הרשות",
    "מאפיינים מרכזיים של אוכלוסייה ברשות",
    "אשכול חברתי-כלכלי של הרשות",
    "תחומי ליבה שבהם העומס המשמעותי ביותר כיום",
    "עד כמה מדידת עומס לפי מספר תיקים בלבד משקפת בעיניך את המציאות?",
    "באיזו מידה קיימים אצלכם פערים משמעותיים בין התקינה לבין עומס העבודה בפועל?"
]

# משתני העומס שאותם רוצים להפוך לשורות
value_vars = [
    "עד כמה המשתנים בשאלות הבאות משקפים עומס בפועל בעבודת עו\"ס משפחה?                                      שלב התיק (חדש / פעיל / ותיק)(לפי סולם של גיל התיק במחלקה)",
    "רמת אינטנסיביות הטיפול ( לפי מספר העדכונים בתיק בחודש)",
    "מספר בני המשפחה המעורבים (לפי מס׳ בני המשפחה שנכללים בתיק)",
    "ריבוי בעיות ותחומי קושי(לפי מספר המאפיינים בתיק)",
    "עבודה חוקית / רגולטורית ( האם יש מעורבות עו״ס לחוק)",
    "עבודה בין מערכתית ( האם נדרשה וותת״ט ולפי מס גורמים נוספים שיש בתיק)"
]

# הפיכה מ-wide ל-long
df_long = df.melt(
    id_vars=id_vars,
    value_vars=value_vars,
    var_name="משתנה",
    value_name="ערך משתנה"
)

# ניקוי שם המשתנה הראשון הארוך
df_long["משתנה"] = df_long["משתנה"].str.replace(
    "עד כמה המשתנים בשאלות הבאות משקפים עומס בפועל בעבודת עו\"ס משפחה?",
    "",
    regex=False
).str.strip()

renames = {
 'מאפיינים מרכזיים של אוכלוסייה ברשות': 'מאפייני אוכלוסייה',
 'אשכול חברתי-כלכלי של הרשות': 'אשכול חברתי-כלכלי',
 'תחומי ליבה שבהם העומס המשמעותי ביותר כיום': 'תחומי העומס',
 'עד כמה מדידת עומס לפי מספר תיקים בלבד משקפת בעיניך את המציאות?': 'דיוק מדידה',
 'באיזו מידה קיימים אצלכם פערים משמעותיים בין התקינה לבין עומס העבודה בפועל?': 'פער מול תקינה'}

df_long = df_long.rename(renames, axis=1)
# שמירה
df_long.to_excel("survey_long_format.xlsx", index=False)
```
